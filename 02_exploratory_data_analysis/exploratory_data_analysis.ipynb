### Функции которые использовал больше двух раз в проекте

```python
def my_describe(column):
    '''
    Функция принимает название столбца.
    Выдает результат describe() только с удобным для визуального анализа форматом. 
    '''
    return df[column].describe().apply('{:,}'.format)
```

```python
def my_abnormal(column, number):
    '''
    Функция принимает называние метрики (column) и значение с которым надо сравнивать.
    Возвращает записей >= или <= указанного значения и их % в данной метрике
    '''
    total = len(df[column])
    
    count_up = len(df[df[column] >= number])
    percent_up = count_up / total
    
    count_down = len(df[df[column] <= number])
    percent_down = count_down / total
    
    
    
    return print('Количество {} в DataFrame\n'
                 '>= {:,}: {}. Доля таких значений {:.2%}\n'
                 '<= {:,}: {}. Доля таких значений {:.2%}'.format(column, number, count_up, 
                                                               percent_up, number, count_down, percent_down))
```

```python
def my_nan_count(name):
    '''
    Функция принимает название столбца.
    Возвращает количество NaNов
    '''
    nan = df[name].isna().sum()
    if nan > 0:
        return print('Метрика {} содержит {} NaN. Надо разбираться.'.format(name, nan))
    else:
        return print('Метрика {} не содержит NaN или они уже были обработаны'.format(name))
```

```python
def group_area_median(column):
    '''
    Функция принимает название столбца и меняет в нем NAN на median() в зависимости от количества комнат в квартире
    '''
    
    # Собираю в список колличество комнат в квартире с пустынми значениями
    rooms_area_empty = list(df[(df[column].isna())]['rooms'].unique())
    rooms_area_empty.sort()
    rooms_area_empty
    
    
    for i in rooms_area_empty:
        df.loc[((df['rooms']==i)&(df[column].isna()==True)), column] = \
        df[df['rooms']==i][column].median()
```

Google выдает подсказку, что в действующих СНиПах указана лишь минимальная высота потолков, ниже которых строительство запрещено - это 2.5 метра. А здравый смысл подсказывает, что потолок выше 5 метров это как-то не разумно для квартиры. А уж 10-100 метров так вообще фантактиска. Считаю, что тут явная ошибка ввода. Например пользователь хотел ввести 2.50 метра, но ошибся и указал 25.0 метров. Таких данных не много и их легок спасти.

```python
def clear_ceiling_height(row):
    '''
    Функция принимает строку DataFrame.
    Если высота потолка ниже 2.5м, то значение меняется на минимально допустимое (2.5)
    Если высота потолка от 5.5м и до 10м, то значение меняется на минимально допустимое (2.5)
    Если высота потолка ниже 2.5м, то значение меняется на минимально допустимое (2.5)
    Если высота потолка в диапазоне от 2.5м до 5.5м, то значение не меняется
    
    '''
    target = row['ceiling_height_round']

    # Обработка значений > 10
    if target >= 10:
        result = round(target / 10, 1)
        # Так как в df есть высота потолка 100м, и это однакомнатная квартира, то
        # добавил дополнительную обработку. Скорее всего высота потолка в этой квартире идет
        # по минимальному стандарту в 2.5 
        # В рамках функции заменю высоту на 1, так как все, что ниже 2.5 буду обрабатывать далее
        if result == 10:
            return 1
        else:
            return result    
    
    # Обработка значений от 5.5 до 10
    elif 5.5 < target < 10:
            return 2.5
    
    # Обработка значений < 2.5
    elif target < 2.5:
        return 2.5
        
    # Если все корректно, то оставляем текущее значение
    else:
        return target
```

```python
df['ceiling_height_round'] = df.apply(clear_ceiling_height, axis=1)
```

Стало лучше, но для дальнейшего удобства работы с высотой потолков сделаю группировку.

```python
def group_ceiling_height(row):
    target = row['ceiling_height_round']
    
    if target < 2.5:
        return 2.5
    elif 2.5 < target < 3:
        return 3
    elif 3 < target < 4.0:
        return 4.0
    elif 4.0 < target:
        return 4.5
    else:
        return target
```

### этаж квартиры; варианты — первый, последний, другой

```python
def my_floor_type(row):
    '''
    Функция принимает строку DataFrame и вытаскивает из нее этажность дома, этаж квартиры.
    Возвращает тип этажа: первый, последний, дургой.
    '''
    floors_total = row['floors_total']
    floor = row['floor']
    
    if floor == 1:
        return 'первый'
    elif floor == floors_total:
        return 'последний'
    else:
        return 'другой'
```

```python
def my_boxplot(df):
    '''
    Функция строит графики по фиксированным переменным
    '''  
    target = ['last_price','total_area']
    
    for i in target:
        fig, ax = plt.subplots(figsize=(15, 3))
        df[i].plot(kind='box')
    return plt.show()
```

Очень хорошо видно, что есть аномально высокие значения. Так как «усы» отображают весь разброс точек кроме выбросов, то рассмотрю значения "усов" более детально.

```python
def my_clear_abnormal(df, q1, q3, coeff):
    '''
    Функция вычисляет и выводит границы "усов" для заданных (фиксированно) переменных
    значения 1го, 3го квартиля и коэффициента для расчета IQR нужно передавать аргументами. 
    '''  
    target = ['last_price','total_area', 'rooms']
    
    for i in target:
        Q1 = df[i].quantile(q1)
        Q3 = df[i].quantile(q3)
        IQR = Q3 - Q1 
        left = Q1 - (IQR * coeff)
        right = Q3 + (IQR * coeff)
        print('Метрика {}. Левый ус: {:,.0f}. Правый ус: {:,.0f}'.format(i, left, right))
```

```python
my_clear_abnormal(df, 0.25, 0.75, 1.5)
```
